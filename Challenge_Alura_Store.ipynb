#Importação das bibliotecas

```python
import pandas as pd

```

#Importação dos dados

```python

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do Faturamento

```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

#Imprimindo os resultados
print(f'Faturamento Loja 1: {faturamento_loja1}')
print(f'Faturamento Loja 2: {faturamento_loja2}')
print(f'Faturamento Loja 3: {faturamento_loja3}')
print(f'Faturamento Loja 4: {faturamento_loja4}')

faturamento_total = faturamento_loja1 + faturamento_loja2 + faturamento_loja3 + faturamento_loja4
print(f"\nFaturamento total de todas as lojas: R${faturamento_total:.2f}")
```

#2. Vendas por Categoria

```python
def vendas_por_categoria(loja, numero_da_loja):
    vendas = loja.groupby('Categoria do Produto')['Preço'].sum()
    print(f"\nVendas por categoria na Loja {numero_da_loja}:")
    print(vendas)

# Mostrar resultados para cada loja
vendas_por_categoria(loja, 1)
vendas_por_categoria(loja2, 2)
vendas_por_categoria(loja3, 3)
vendas_por_categoria(loja4, 4)

```

#3. Média de Avaliação das Lojas

```python
#Avaliação
avaliacao_da_compra_loja1 = (loja['Avaliação da compra'].sum())/len(loja)
avaliacao_da_compra_loja2 = (loja2['Avaliação da compra'].sum())/len(loja2)
avaliacao_da_compra_loja3 = (loja3['Avaliação da compra'].sum())/len(loja3)
avaliacao_da_compra_loja4 = (loja4['Avaliação da compra'].sum())/len(loja4)

#Exibindo os resultados
print(f'Média de Avaliação loja 1: {avaliacao_da_compra_loja1:.3}')
print(f'Média de Avaliação loja 2: {avaliacao_da_compra_loja2:.3}')
print(f'Média de Avaliação loja 3: {avaliacao_da_compra_loja3:.3}')
print(f'Média de Avaliação loja 4: {avaliacao_da_compra_loja4:.3}')
```

#4. Produtos Mais e Menos Vendidos

```python

def mais_e_menos_vendidos(lojas, numero):
    quantidade_produtos_loja = lojas['Produto'].value_counts()
    mais_vendidos_loja = quantidade_produtos_loja.head()
    menos_vendidos_loja = quantidade_produtos_loja.tail()

    print(f"\nOs 5 produtos MAIS vendidos da Loja {numero}:\n")
    print(mais_vendidos_loja)

    print(f"\nOs 5 produtos MENOS vendidos da Loja {numero}:")
    print(menos_vendidos_loja)
    print('---------------------------------------------------')

# Mostrar resultados para cada loja
mais_e_menos_vendidos(loja, 1)
mais_e_menos_vendidos(loja2, 2)
mais_e_menos_vendidos(loja3, 3)
mais_e_menos_vendidos(loja4, 4)

```

#Frete Médio por Loja

```python
frete_medio_por_loja = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Frete médio por loja': [
        loja['Frete'].mean(),
        loja2['Frete'].mean(),
        loja3['Frete'].mean(),
        loja4['Frete'].mean()
    ]
})

print(frete_medio_por_loja)
```

#Relatório Final

##**Objetivo:**

Identificar qual das quatro lojas da rede Alura Store apresenta pior desempenho, com base em métricas de faturamento, vendas por categoria, avaliação dos clientes, produtos mais/menos vendidos e frete médio, e recomendar qual loja o Sr. João deve vender.


##**Introdução**
O Sr. João possui quatro lojas virtuais (Loja 1, Loja 2, Loja 3 e Loja 4) no e-commerce Alura Store. O objetivo desta análise é comparar o desempenho das lojas por meio de métricas comerciais e de experiência do cliente, para recomendar qual loja é a melhor candidata para venda — ou seja, qual loja apresenta menor desempenho e menor potencial de recuperação, liberando capital para novos investimentos.


##**Metodologia**
Foram lidos os arquivos CSV correspondentes às quatro lojas e realizadas análises separadas por loja.

Métricas calculadas:

- Faturamento total (soma da coluna Preço).

- Vendas por categoria (contagem por Categoria do Produto).

- Média de avaliação dos clientes (média da coluna Avaliação da compra).

- Produtos mais e menos vendidos (value_counts por Produto).

- Frete médio (média da coluna Frete).

Também foram gerados gráficos para facilitar a visualização.


Tratamento: utilizadas as colunas originais conforme entregues; pressupõe-se consistência dos nomes de colunas e ausência de valores inválidos significativos.


##**Conclusão e recomendação**

Com base nas métricas avaliadas:

Recomendação principal: vender a Loja 4.

**Justificativas:**

1. Menor faturamento absoluto entre as quatro lojas — métrica crítica para liberar capital.

2. Volume de vendas por produto é inferior ao das outras lojas.

3. Apesar de ter frete médio menor (ponto positivo), isso não se refletiu em maior faturamento.

4. A média de avaliação da Loja 4 não compensa o baixo desempenho em receita.

Diante de tais constatações após a análise das quatro lojas da Alura Store (faturamento, categorias, avaliações, produtos e frete), recomenda-se a venda da Loja 4, que apresenta o menor faturamento (R$ 3.591,87) e volumes de venda por produto inferiores aos das demais. A decisão final deve considerar margem de lucro e custos operacionais, mas com base nos dados disponíveis, a Loja 4 é a candidata mais apropriada para venda.

**Observação:**  a decisão final deve considerar também fatores não presentes nos dados analisados: margem de lucro por produto, custo operacional da loja, campanhas de marketing ou sazonalidade, contratos/fornecedores, localização do estoque e potenciais planos de recuperação (ex.: promoções, reposicionamento de mix). Se houver possibilidade de aumentar receita com ações concretas de curto prazo e baixo custo, pode valer a pena reconsiderar.
